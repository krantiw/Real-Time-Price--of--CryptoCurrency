##### Create a Function to get the price of Cryptocurrency

```python
# div class="BNeawe iBp4i AP7Wnd"

def get_price(coin):
    url = "https://www.google.com/search?q="+coin+"+price"
    
    #Make a request to the website
    html = requests.get(url)
    
    # Parse the html
    soup = BeautifulSoup(html.text, "html.parser")
    
    #find the current price
    text1 = soup.find("div" , attrs = {"class" : "BNeawe iBp4i AP7Wnd"}).find("div" , attrs = {"class" : "BNeawe iBp4i AP7Wnd"}).text
    
    #return the text
    return text1
```

```python
#test the function
# Get the price of cryptocurreny

litecoin = get_price("litecoin")
bitcoin = get_price("bitcoin")
ethereum = get_price("ethereum")


print("The Real Time Price of bitcoin is : ", bitcoin)
print("The Real Time Price of litecoin is : ", litecoin)
print("The Real Time Price of ethereum is : ", ethereum)
```

##### Create a function to consistently show the price of bitcoin when it changes

```python
def change():
    last_price = -1
    #create a loop to continously show the price
    while True:
        #choose the cryptocurrency
        crypto = "bitcoin"
        #get the current price of crypto
        current_price = get_price("bitcoin")
        
        #Check if the price has changed
        if current_price != last_price:
            print(crypto+ " price:", current_price)
            last_price = current_price
        time.sleep(3)
```

##### Execute the change function

```python
change()
```
